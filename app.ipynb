Inicializando o driver

```python
# Configurações do Chrome
options = Options()
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
prefs = {
    "profile.managed_default_content_settings.images": 2,  # Bloqueia imagens
    "profile.managed_default_content_settings.stylesheets": 2,  # Bloqueia CSS
    "profile.managed_default_content_settings.fonts": 2  # Bloqueia fontes
}
options.add_experimental_option("prefs", prefs)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.minimize_window()


def generate_id(*args):
    """Gera um ID único baseado nos argumentos fornecidos."""
    string_to_hash = "_".join(map(str, args))
    return hashlib.md5(string_to_hash.encode()).hexdigest()[:8]  # Retorna os primeiros 8 caracteres do hash
```

Puxando os dados do JSON e pegando os times de cada liga

```python
with open("api.json", "r", encoding="utf-8") as f:
    dados_ligas = json.load(f)
```

```python
for pais, info in dados_ligas["Paises"].items():
    print(f'Processando país: {pais}')
    
    # Itera por cada liga no país
    for liga in info["ligas"]:
        nome = liga["Nome da Liga"]
        link = liga["Link da Liga"]
        qnt = liga["Quantidade de equipes"]
        qnt = int(qnt)
        driver.get(link)
        
        print(f'Abrindo liga: {nome} ({link})')
        for i in range(1, qnt+1):
            link_equipe = driver.find_element('xpath', f'/html/body/div[1]/main/div[1]/div[1]/div[2]/div[2]/div/table/tbody/tr[{i}]/td[2]/a[1]').get_attribute('href')
            nome_equipe = driver.find_element('xpath', f'/html/body/div[1]/main/div[1]/div[1]/div[2]/div[2]/div/table/tbody/tr[{i}]/td[2]/a[1]').get_attribute('title')
            qnt_jogadores = driver.find_element('xpath', f'/html/body/div[1]/main/div[1]/div[1]/div[2]/div[2]/div/table/tbody/tr[{i}]/td[3]/a').get_attribute('innerText')
            id_time = generate_id(nome_equipe, nome)
            if nome_equipe not in liga["times"]:
                liga["times"][nome_equipe] = {
                    "id": id_time,
                    "Link da Equipe": link_equipe,
                    "Quantidade de Jogadores": qnt_jogadores,
                    "Jogadores": {}
                }
                print(f"Equipe '{nome_equipe}' adicionada com sucesso!")
            else:
                print(f"Equipe com ID '{id_time}' - {nome_equipe} já existe na liga.")
                liga["times"][nome_equipe]["Link da Equipe"] = link_equipe
                liga["times"][nome_equipe]["Quantidade de Jogadores"] = qnt_jogadores
                print(f'Time atualizado: {id_time} - {nome_equipe} ({qnt_jogadores})')
        with open("api.json", "w", encoding="utf-8") as f:
            json.dump(dados_ligas, f, ensure_ascii=False, indent=4)
```

Puxando as informações dos jogadores de cada time

```python
x = 2

for pais, info_pais in dados_ligas["Paises"].items():
    print(f'Processando país: {pais}')
    
    # Itera por cada liga no país
    for liga in info_pais["ligas"]:
        print(f'Processando liga: {liga["Nome da Liga"]}')

        for time, info_time in liga["times"].items():
            if info_time["Jogadores"] and x == 2:
                print(f"Time '{time}' já possui jogadores cadastrados. Pulando...")
                continue
            print(f'Processando time: {time} ({qnt_jogadores})')
            link = info_time["Link da Equipe"]
            driver.get(link)
            qnt = info_time["Quantidade de Jogadores"]
            qnt = int(qnt)

            for i in range(1, qnt+1):
                try:
                    nome_jogador = driver.find_element('xpath', f'/html/body/div[1]/main/div[1]/div[1]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[2]/table/tbody/tr[1]/td[2]/a').get_attribute('innerText')
                    link_jogador = driver.find_element('xpath', f'/html/body/div[1]/main/div[1]/div[1]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[2]/table/tbody/tr[1]/td[2]/a').get_attribute('href')
                    posicao = driver.find_element('xpath', f'/html/body/div[1]/main/div[1]/div[1]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[2]/table/tbody/tr[2]/td').get_attribute('innerText')
                    dtnasc = driver.find_element('xpath', f'/html/body/div[1]/main/div[1]/div[1]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[3]').get_attribute('innerText')
                    nacionalidade = driver.find_element('xpath', f'/html/body/div[1]/main/div[1]/div[1]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[4]/img').get_attribute('title')
                    id_player = generate_id(nome_jogador, dtnasc)
                    try:
                        valor = driver.find_element('xpath', f'/html/body/div[1]/main/div[1]/div[1]/div[1]/div[3]/div/table/tbody/tr[{i}]/td[5]/a').get_attribute('innerText')
                    except:
                        valor = 0
                    
                    if nome_jogador not in info_time["Jogadores"]:
                        info_time["Jogadores"][nome_jogador] = {
                            "id": id_player,
                            "Link do Jogador": link_jogador,
                            "Posição": posicao,
                            "Nacionalidade": nacionalidade,
                            "Idade": dtnasc,
                            "Valor de Mercado": valor,
                            "Estatísticas": {}
                        }
                        print(f"Jogador '{nome_jogador}' adicionado com sucesso!")
                    else:
                        print(f"Jogador com ID '{id_player}' - {nome_jogador} já existe no time.")
                        info_time["Jogadores"][nome_jogador].update({
                            "Link do Jogador": link_jogador,
                            "Posicao": posicao,
                            "Nacionalidade": nacionalidade,
                            "Valor de Mercado": valor
                        })   
                        print(f'Jogador atualizado: {id_player} - {nome_jogador} ({posicao}) - {nacionalidade} ({dtnasc} anos) - {valor}')
                except:
                    print(f"Erro ao processar jogador - {id_player} - {nome_jogador}")
                    with open("api.json", "w", encoding="utf-8") as f:
                        json.dump(dados_ligas, f, ensure_ascii=False, indent=4)
                    continue
            with open("api.json", "w", encoding="utf-8") as f:
                json.dump(dados_ligas, f, ensure_ascii=False, indent=4)
```
